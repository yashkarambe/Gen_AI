```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python

# Load a pre-trained model for code generation (such as GPT models fine-tuned on code)
model_name = "codellama/CodeLlama-7b-Instruct-hf"  # This model is trained on code datasets
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def generate_code(prompt, max_length=200):                                  #--> Function to generate code based on input prompt
    inputs = tokenizer(prompt, return_tensors="pt")                         #--> Tokenize the input prompt
    output = model.generate(**inputs, max_length=max_length, 
            num_beams=5, no_repeat_ngram_size=2, early_stopping=True)       #--> Generate code using the model
    generated_code = tokenizer.decode(output[0], skip_special_tokens=True)  #--> Decode the generated tokens into code string
    return generated_code
```

```python
prompt = "Write a Python function to sort a list of integers using sellection sort."   
generated_code = generate_code(prompt)
print("Generated Code:\n")
print(generated_code)
```

```python
prompt = "Write a Python function to check number is prime or not"
generated_code = generate_code(prompt)
print("Generated Code:\n")
print(generated_code)
```

```python
def fibonacci(n):
    if n <= 0:
        return "Input should be a positive integer."
    elif n == 1:
        return 0
    elif n == 2:
        return 1
    else:
        a, b = 0, 1
        for _ in range(2, n):
            a, b = b, a + b
        return b

# Example usage
n = 5
print(f"Fibonacci of {n} is: {fibonacci(n)}")
```
